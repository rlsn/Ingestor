```python
import ingestor
```

```python
# Show available datasets
dataset_names = ingestor.list_datasets(display = True)
```

```python
# load the dataset, automatically download necessary files if the they are not already there
ds = ingestor.load_dataset("wikimedia/wikipedia", "20231101.ch", downloaded_only=False)
ds
```

```python
# The above process can be broken separate stages as well.
# Download the subset "20231101.ady" from "wikimedia/wikipedia"
ingestor.download("wikimedia/wikipedia", "20231101.ady", force_redownload=True)
# Load the subset into memory
ds = ingestor.load_dataset("wikimedia/wikipedia", "20231101.ady", downloaded_only=True)
ds
```

```python
# view the subset info of a dataset
subsets = ingestor.list_subsets("wikimedia/wikipedia")
```

```python
# Download specific partitions only
partitions = ingestor.list_partitions("wikimedia/wikipedia", "20231101.en", display=False)[:2]
ingestor.download("wikimedia/wikipedia", "20231101.en", partitions=partitions, force_redownload=False)
```

```python
# load the downloaded partitions, which are automatically joined together
ds = ingestor.load_dataset("wikimedia/wikipedia", "20231101.en", downloaded_only=True)
ds
```

```python
# Alternatively, partitions can be specified
partitions = ingestor.list_partitions("wikimedia/wikipedia", "20231101.en", display=False)[:1]
ds = ingestor.load_dataset("wikimedia/wikipedia", "20231101.en", partitions=partitions, downloaded_only=False,)
ds
```

```python
# View partition info in a subset
partitions = ingestor.list_partitions("wikimedia/wikipedia", "20231101.en")
```

```python
# datasets can be updated to latest version by forcing redownload
ingestor.download("wikimedia/wikipedia", "20231101.ady", force_redownload=True)

# remove a dataset or subset to free storage
ingestor.remove("wikimedia/wikipedia", "20231101.ady", force_remove=True)
```
